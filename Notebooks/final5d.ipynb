## Guided Capstone Report

We set out to use market data from ski resorts all over the country to improve the profitability of Big Mountain resort, specifically to make a strategy to fix Big Mountain Resort's ticket pricing to improve profitabiliy. We aimed to build a predictive model using the ski resort data set, as well as other sources of data like the internet, for the client. We will use the provided data in a CSV file that contains information from 330 resorts in the US that can be considered part of the same market share. Columns of this data include the name of the ski resort, the region, how many trams, how many chairlifts, average snowfall, and more.  

My problem statement was: How can Big Mountain Resort change its ticket pricing strategy to optimize its capitalization on the facilities, as well as cut costs by 10% by the end of the coming season, possibly by introducing ticket tiers for guests using some of the infrastructure, without losing existing business?

We did use the optimization of the capitalization on the facilities in the process, and we can still explore tiered ticketing in the future. Furthermore, we might want to gather data about visitor numbers in the future.

One inital finding from the dataset was that weekend adult tickets were more expensive than weekday. However, the state Big Mountain is in, Montana, has equal ticket prices, so we used only weekend prices in our model because there was less missing data for weekend.



We did a Principal Component Analysis on the data to create a simpler model that covered more variables without being redundant (for example, using two variables like summit and base elevation that are highly correlated) to explain variance. We built models and chose the best one based on how they performed to predict test sets.  It also allowed us to choose a subset of features to give a good model as opposed to using all.  By trying different numbers k of features, we found k=8 was optimal. Throughout the process we used cross validation - cycling through test sets by how many partitions we had, so that our models were tested all throughout the data instead of just on one test set. The linear model found that vertical drop and snow making area covered were the most important estimators. We tried a random forest regressor, too. The important features for RF were also vertical drop, snow making area covered, as well as fastQuads and Runs. The RF model performed better in CV mean absolute error, so we used that!

We used the model to predict an appropriate ticket price for Big Mountain. Big Mountain currently charges 81. Our best model predicts that a price consistent with the market, based on Big Mountain’s features, can be 95.87. Even with the expected mean absolute error of $10.39, this suggests there is room for an increase. Big Mountain’s features suggest that a higher price would work.
